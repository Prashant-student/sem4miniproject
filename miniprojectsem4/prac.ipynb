```python
c_name = data.CROP_TYPE_NAME.unique()
```

```python
gsm = []
for i in range(0, len(c_name)):
    gsm.append((len(data[(data['CROP_TYPE_NAME'] == c_name[i]) & (data['DIVERSION_CENTER_NAME'] == 'GSM_HARVEST')].CROP_TYPE_NAME) / len(
        data[data['CROP_TYPE_NAME'] == c_name[i]].CROP_TYPE_NAME)) * 100)
```

```python
plt.bar(c_name, gsm)
plt.xlabel('CROP_TYPE')
plt.ylabel('CORPS TAKEN BY GSM_HARVEST IN %')
plt.show()
```
